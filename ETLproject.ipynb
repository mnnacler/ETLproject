```python
# SQLAlchemy
from sqlalchemy import create_engine, Column, Integer, String, Float, Date
from sqlalchemy.ext.declarative import declarative_base
import pandas as pd
import numpy as np
# Path to sqlite

Base = declarative_base()

```

```python
netflixdf = pd.read_csv("netflix_titles.csv")
netflixdf = netflixdf.set_index('show_id')
del netflixdf['description']
```

```python
bbDf = pd.read_csv("Blockbusters_2019-1977.csv")
bbDf = bbDf.set_index('film_title')
bbDf
```

```python
oscarDf = pd.read_csv("the_oscar_award.csv")
oscarDf["ID"] = np.arange(oscarDf.shape[0])
oscarDf = oscarDf.set_index('ID')
oscarDf
```

```python
class netflix(Base):
    __tablename__ = "netflix"
    show_id = Column(Integer, primary_key = True)
    type = Column(String(255))
    title = Column(String(255))
    director = Column(String(255))
    cast = Column(String(255))
    country = Column(String(255))
    date_added = Column(Date)
    release_year = Column(Integer)
    rating = Column(String(255))
    duration = Column(String(255))
    listed_in = Column(String(255))
```

```python
class blockbuster(Base):
    __tablename__="blockbusters"
    release_year= Column(Integer)
    rank_in_year = Column(Integer)
    imdb_rating=Column(Integer)
    mpaa_rating=Column(String(255))
    film_title=Column(String(255), primary_key=True)
    film_budget=Column(String(255))
    length_in_min=Column(Integer)
    domestic_distributor=Column(String(255))
    worldwide_gross=Column(String(255))
    domestic_gross=Column(String(255))
    genre_1=Column(String(255))
    genre_2=Column(String(255))
    genre_3=Column(String(255))
```

```python
class oscars(Base):
    __tablename__ = "oscars"
    ID = Column(Integer, primary_key = True)
    year_film = Column(Integer)
    year_ceremony = Column(Integer)
    ceremony = Column(Integer)
    category = Column(String(255))
    name = Column(String(255))
    film = Column(String(255))
    winner = Column(String(255))
```

```python
database_path = "postgresql://postgres:postgres@localhost:5432/ETL"
engine = create_engine(database_path)
conn = engine.connect()
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session
session = Session(bind = engine)
```

```python
netflixdf.to_sql('netflix', con=engine, if_exists = 'append')
bbDf.to_sql('blockbusters', con=engine, if_exists = 'append')
oscarDf.to_sql('oscars', con=engine, if_exists = 'append')
```
